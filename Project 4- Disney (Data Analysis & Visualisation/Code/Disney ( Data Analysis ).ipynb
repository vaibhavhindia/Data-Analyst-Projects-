# Convert Date Time format

```python
df["date_added"] = pd.to_datetime(df['date_added'])
df['day_added'] = df['date_added'].dt.day
df['year_added'] = df['date_added'].dt.year
df['month_added']=df['date_added'].dt.month
df['year_added'].astype(int);
df['day_added'].astype(int);
```

# Data Visualisation 

# Type : Movie & Tv shows

```python
sns.countplot(disney['type'])
fig=plt.gcf()
fig.set_size_inches(10,10)
plt.title('Type')
```

# Rating of Shows and Movies

```python
sns.countplot(disney['rating'])
sns.countplot(disney['rating']).set_xticklabels(sns.countplot(disney['rating']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(13,13)
plt.title('Rating')
```

# Now Check the Relation Between Type and Rating

```python
plt.figure(figsize=(16,8))
sns.countplot(x='rating',hue='type',data=disney)
plt.title('Relation between Type and Rating')
plt.show()

```

# Pie-chart for the Type: Movie and TV Shows

```python
labels = ['Movie', 'TV show']
size = disney['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

# Pie-chart for Rating

```python
disney['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(18,12))
plt.show()

```
